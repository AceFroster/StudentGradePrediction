**Old Version**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score
from imblearn.over_sampling import SMOTE
```

```python
X = df.drop(columns=['Grade'])
y = df['Grade']
```

```python
cate_feature = ['Gender', 'Department', 'Extracurricular_Activities', 'Parent_Education_Level']
num_feature = ['Age', 'Attendance (%)', 'Midterm_Score', 'Final_Score', 'Assignments_Avg', 'Quizzes_Avg',
               'Participation_Score', 'Projects_Score', 'Study_Hours_per_Week']
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_feature),
        ('cat', OneHotEncoder(), cate_feature)
    ])
```

```python
X_preprocessed = preprocessor.fit_transform(X)
```

```python
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_preprocessed, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size = 0.2, random_state = 42, stratify = y_resampled)
```

```python
model = Pipeline(steps = [
    ('classifier', RandomForestClassifier(random_state=42))
])
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print("Classification Report: ", classification_report(y_test, y_pred))
print("ROC-AUC Score: ", roc_auc_score(y_test, model.predict_proba(X_test), multi_class='ovr'))
```

**Improved Version**

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'classifier__n_estimators': [100, 200, 300],
    'classifier__max_depth': [None, 10, 20],
    'classifier__min_samples_split': [2, 5, 10],
    'classifier__min_samples_leaf': [1, 2, 4],
}

grid_search = GridSearchCV(model, param_grid, cv = 5, scoring='roc_auc_ovr')
grid_search.fit(X_train, y_train)
```

```python
better_model = grid_search.best_estimator_
y_pred = better_model.predict(X_test)
```

```python
print("Classification Report: ", classification_report(y_test, y_pred))
print("ROC-AUC score: ", roc_auc_score(y_test, better_model.predict_proba(X_test), multi_class = 'ovr'))
```
